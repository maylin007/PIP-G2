```python
def trackObject(x1,y1,r1,x2,y2,r2,x3,y3,r3):
  A = 2*x2 - 2*x1
  B = 2*y2 - 2*y1
  C = r1**2 - r2**2 - x1**2 + x2**2 - y1**2 + y2**2
  D = 2*x3 - 2*x2
  E = 2*y3 - 2*y2
  F = r2**2 - r3**2 - x2**2 + x3**2 - y2**2 + y3**2

  x = (C*E - F*B) / (E*A - B*D)
  y = (C*D - A*F) / (B*D - A*E)
  return x,y
```

```python
import random

df2 = pd.DataFrame(columns=['seqnum1', 'x1', 'y1', 'r1', 'seqnum2', 'x2', 'y2', 'r2', 'seqnum3', 'x3', 'y3', 'r3', 'xTrue', 'yTrue', 'xPred', 'yPred'])

for i in range(len(df)):
    
    if (i%6 == 0):
        
        option1, option2, option3 = random.sample(range(i+1, i+7), 3)
        OK = False
        while OK == False:
            try : 
                r1 = int(df[df.seqnum == option1].distance)
                r2 = int(df[df.seqnum == option2].distance)
                r3 = int(df[df.seqnum == option3].distance)
                OK = True
            except:
                option1, option2, option3 = random.sample(range(i+1, i+7), 3)

        x1 = int(df[df.seqnum == option1].loc_target_x)
        y1 = int(df[df.seqnum == option1].loc_target_y)  
        x2 = int(df[df.seqnum == option2].loc_target_x)
        y2 = int(df[df.seqnum == option2].loc_target_y)  
        x3 = int(df[df.seqnum == option3].loc_target_x)
        y3 = int(df[df.seqnum == option3].loc_target_y) 
        xTrue = int(df[df.seqnum == option1].loc_initiator_x)
        yTrue = int(df[df.seqnum == option1].loc_initiator_y)
        
        xPred, yPred = trackObject(x1,y1,r1,x2,y2,r2,x3,y3,r3)
        
    df2 = df2.append({'x1':x1, 'y1':y1, 'r1':r1, 
                      'x2':x2, 'y2':y2, 'r2':r2, 
                      'x3':x3, 'y3':y3, 'r3':r3, 
                      'seqnum1': option1, 'seqnum2': option2, 'seqnum3': option3,
                      'xTrue':xTrue, 'yTrue':yTrue, 
                      'xPred':xPred, 'yPred':yPred} , ignore_index=True)

df2.drop_duplicates(keep = 'first', inplace=True)
```
