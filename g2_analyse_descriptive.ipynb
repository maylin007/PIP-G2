## Correlation between variables

```python
data_corr = data.corr()
corr=pd.DataFrame(data_corr)
c = ["darkred","red","lightcoral","white", "palegreen","green","darkgreen"]
v = [0,.15,.4,.5,0.6,.9,1.]
l = list(zip(v,c))
GnRd=LinearSegmentedColormap.from_list('rg',l, N=256)
plt.figure(1,figsize = (25, 20))
ax = sns.heatmap(data.corr(), annot=True,cmap=GnRd)
```

```python
isolated = []
for nom in data_corr.columns :
    correl = []
    for i in range(len(data_corr.columns)):

        if abs(data_corr[nom][i]) > 0.6 and nom != data_corr.columns[i] :
            correl.append(data_corr.columns[i])
    if correl != []:
        print('In correlation with ',nom,' there are :', correl,'\n')
    else:
        print("No variables are corelated with ", nom, '\n')
        isolated.append(nom)
data_corr_clear = data_corr.drop(isolated,axis=1) #to have the correlation's matrix with corralted values
data_corr_clear = data_corr_clear.drop(isolated,axis=0)
```

```python
#Matrix of correlations with variables correlated
plt.figure(1,figsize = (25, 20))
ax = sns.heatmap(data_corr_clear, annot=True,cmap=GnRd)
```

t1,t2,t3,t4 are correlated. This is not surprising. Same for rssi columns. tof and range are correlated and this is logical because it's the definition of speed. The surprising correlations are between ditances and temperature and positions.

## Histogramm of quantitative data

```python
for column in data.select_dtypes(['number']).columns :
    sns.displot(data=data[column], kde=True)
    plt.title("histogram of "+str(column))
    plt.show()
```

Ranging_error appears to be Gaussian. We will do a test to verify

```python
#importation of librarie needed for normality test
from scipy import stats
```

```python
shapiro_test = stats.shapiro(data['ranging_error'])
print('pvalue associated ', shapiro_test.pvalue)
```

This pvalue is <<0.05. It cannot be assumed that the distance estimation error is normally distributed.
